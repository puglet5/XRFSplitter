```python
def sorter(x, y):
    strx: str = str(x[1]).strip()
    stry: str = str(y[1]).strip()

    try:
        x = float(strx)
        y = float(stry)

        if x < y:
            return -1
        elif x > y:
            return 1
        else:
            return 0

    except ValueError:
        x = strx
        y = stry

    if x == "" and y == "< LOD":
        return -1
    if y == "" and x == "< LOD":
        return 1
    if x == "" or x == "< LOD" and y != x:
        return -1
    if y == "" or y == "< LOD" and x != y:
        return 1
    if x < y:
        return -1
    elif x > y:
        return 1
    else:
        return 0


key = functools.cmp_to_key(sorter)
```
